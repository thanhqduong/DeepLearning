```python
import cv2
import matplotlib.pyplot as plt
import time
%matplotlib inline
```

```python
cv2.__version__
```

```python
def convertToRGB(img):
  return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def detect(f_cascade, colored_img, scaleFactor = 1.25):
  img_copy = colored_img.copy()
  gray = cv2.cvtColor(img_copy, cv2.COLOR_BGR2GRAY)
  nums = f_cascade.detectMultiScale(gray, scaleFactor=scaleFactor,
  minNeighbors=5)
  found = len(nums)
  for (x, y, w, h) in nums:
    cv2.rectangle(img_copy, (x, y), (x+w, y+h), (0, 255, 0), 2)
  return img_copy, found
```

```python
glasses_cascade = cv2.CascadeClassifier('haarcascade_eye_tree_eyeglasses.xml')
```

```python
body_cascade = cv2.CascadeClassifier('haarcascade_fullbody.xml')
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
test1 = cv2.imread('face_glasses_1.jpg')
img, found = detect(glasses_cascade, test1, scaleFactor = 1.01)
plt.imshow(convertToRGB(img))
```

```python
img, found = detect(face_cascade, img, scaleFactor = 1.01)
plt.imshow(convertToRGB(img))
```

```python
test2 = cv2.imread('body.jpg')
img, found = detect(body_cascade, test2)
plt.imshow(convertToRGB(img))
```
